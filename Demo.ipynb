```python
df_q = pd.read_csv("learningData_wikify.csv",index_col = False)
df_q = df_q.drop(["Unnamed: 0"], axis=1)
```

```python
# Link : https://www.wikidata.org/wiki/
df_q['FIPS'] = df_q['FIPS'].astype('str') 
for idx, rows in df_q.iterrows():
    if rows['FIPS_wikidata'] is not np.nan:
        rows['FIPS'] = str(rows['FIPS']) + '#https://www.wikidata.org/wiki/' + str(rows['FIPS_wikidata'])
        df_q.at[idx,'FIPS'] = rows['FIPS']
    else:
        df_q.at[idx,'FIPS'] = str(rows['FIPS'])
    #print(idx,rows['FIPS'])
```

```python
df_q['Area'] = df_q['Area'].astype('str') 
for idx, rows in df_q.iterrows():
    if rows['FIPS_wikidata'] is not np.nan:
        rows['Area'] = str(rows['Area']) + '#https://www.wikidata.org/wiki/' + str(rows['FIPS_wikidata'])
        df_q.at[idx,'Area'] = rows['Area']
    else:
        df_q.at[idx,'Area'] = str(rows['Area'])
```

```python
df_q['State'] = df_q['State'].astype('str') 
for idx, rows in df_q.iterrows():
    if rows['State_wikidata'] is not np.nan:
        rows['State'] = str(rows['State']) + '#https://www.wikidata.org/wiki/' + str(rows['State_wikidata'])
        df_q.at[idx,'State'] = rows['State']
    else:
        df_q.at[idx,'State'] = str(rows['State'])
```

```python
def make_clickable_both(val): 
    if '#' not in val:
        return val
    name, url = val.split('#')
    return f'<a href="{url}">{name}</a>'
df_q.loc[:10].style.format({'FIPS': make_clickable_both, 'State' : make_clickable_both, 'Area' : make_clickable_both})
```
